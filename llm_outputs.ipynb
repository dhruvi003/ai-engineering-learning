```python
import json
```

```python
def extract_person_raw(text: str) -> dict:
    response = raw_client.chat.completions.create(
        model=MODEL,
        messages=[
            {
                "role": "system",
                "content": "You are a data extractor. Always respond with valid JSON only. No explanation, no markdown."
            },
            {
                "role": "user",
                "content": f"Extract name, age, and city from this text. Return JSON with keys: name, age, city.\n\nText: {text}"
            }
        ],
        temperature=0,
        response_format={"type": "json_object"}  # forces JSON output
    )

    return response
```

```python
# MODEL = "llama3-8b-8192"
result = extract_person_raw("Priya is 28 years old and lives in Bangalore.")
```

```python
print(json.dumps(result.model_dump(), indent=2))
```

```python
class PersonExtract(BaseModel):
    name: str = Field(description="Full name of the person")
    age: int = Field(description="Age as a whole number")
    city: str = Field(description="City of residence")
```

```python
def extract_person(text: str) -> PersonExtract:
    return client.chat.completions.create(
        model=MODEL,
        response_model=PersonExtract,  # instructor magic — pass Pydantic model
        messages=[
            {"role": "user", "content": f"Extract person info from: {text}"}
        ],
        temperature=0
    )
```

```python
result = extract_person("Priya is 28 years old and lives in Bangalore.")
print(result)
print(type(result))       # PersonExtract — a real typed Python object
print(result.name)        # "Priya"
print(result.age)         # 28  (integer, not string)
print(result.age + 2)
```

```python
class Sentiment(str, Enum):
    positive = "positive"
    negative = "negative"
    neutral  = "neutral"
```

```python
class ReviewAnalysis(BaseModel):
    sentiment: Sentiment
    confidence: float = Field(ge=0.0, le=1.0, description="Confidence between 0 and 1")
    key_phrases: List[str] = Field(description="Important phrases from the review")
    summary: str = Field(description="One sentence summary")
```

```python
def analyse_review(review: str) -> ReviewAnalysis:
    return client.chat.completions.create(
        model=MODEL,
        response_model=ReviewAnalysis,
        messages=[{"role": "user", "content": f"Analyse this review: {review}"}],
        temperature=0
    )
```

```python
r = analyse_review("Battery life is amazing but the screen is too dim. Overall satisfied.")
print(r.model_dump_json(indent=2))
print(f"\nSentiment: {r.sentiment}")          # Sentiment.positive (enum)
print(f"Sentiment value: {r.sentiment.value}") # "positive" (string)
```

```python
class Address(BaseModel):
    street: Optional[str] = None
    city: str
    country: str
```

```python
class Employee(BaseModel):
    name: str
    role: str
    year_experiece: int
    skills: List[str]
    address: Optional[Address] = None
```

```python
def extract_employee_info(emp_data:str) -> Employee:
    res = client.chat.completions.create(
        model = MODEL,
        response_model= Employee,
        messages= [
            {
                "role":"user",
                "content": f"Extract employee info from: {emp_data}"
            }
        ],
        temperature=0
    )

    return res
```

```python
text = """
Arun Kumar is a senior ML engineer with 6 years of experience.
He knows Python, PyTorch and Docker. Based in Hyderabad, India.
"""
```

```python
class Experience(BaseModel):
    company: str
    role: str
    duration: str
```

```python
class Resume(BaseModel):
    name: str
    email:Optional[str] = None
    skills: List[str]
    experience: List[Experience]
    total_years : Optional[int] = None


```

```python
def parse_resume(text: str) -> Resume:
    res = client.chat.completions.create(
        model = MODEL,
        response_model= Resume,
        messages = [
            {
                "role":"user",
                "content": f"parse the details from the resume given in text {text}"
            }
        ],
        temperature= 0
    )

    return res
```

```python
sample_resume = """
Neha Sharma
neha@email.com

Skills: Python, Machine Learning, TensorFlow, SQL, Docker

Experience:
- Data Scientist at Infosys (2021-2023)
- ML Engineer at Flipkart (2023-present)
"""
```

```python
resume1 = parse_resume(sample_resume)
```
